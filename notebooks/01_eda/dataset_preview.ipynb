# Dataset Preview from data/raw

This notebook loads all CSV files from `data/raw` and shows:
- First 5 rows
- Dataset shape
- Column names

```python
from pathlib import Path
import pandas as pd
```

```python
data_dir = Path('../../data/raw')
csv_files = sorted(data_dir.glob('*.csv'))

if not csv_files:
    print('No CSV files found in data/raw')
else:
    for csv_path in csv_files:
        print('=' * 80)
        print(f'Dataset: {csv_path.name}')

        df = pd.read_csv(csv_path)

        print('\nFirst 5 rows:')
        print(df.head(5))

        print('\nDataset shape (rows, columns):')
        print(df.shape)

        print('\nColumn names:')
        print(df.columns.tolist())
        print()
```

## Basic EDA Across All Raw Datasets

This cell performs basic exploratory data analysis for every CSV file in `data/raw`:
- `info()`
- `describe()` for numeric columns
- `describe(include='all')`
- Missing value counts and percentages

```python
from pathlib import Path
import pandas as pd

raw_dir = Path('../../data/raw')
raw_csv_files = sorted(raw_dir.glob('*.csv'))

if not raw_csv_files:
    print('No CSV files found in data/raw')
else:
    for file_path in raw_csv_files:
        print('\n' + '=' * 90)
        print(f'Dataset: {file_path.name}')
        print('=' * 90)

        df = pd.read_csv(file_path)

        print('\n1) Dataset Info')
        df.info()

        print('\n2) Descriptive Statistics (Numeric)')
        print(df.describe())

        print('\n3) Descriptive Statistics (All Columns)')
        print(df.describe(include='all'))

        print('\n4) Missing Values (Count)')
        missing_count = df.isnull().sum()
        print(missing_count)

        print('\n5) Missing Values (%)')
        missing_pct = (missing_count / len(df) * 100).round(2)
        print(missing_pct)
```
